```python
import pandas as pd
data = pd.read_csv("loan.csv")
data.head()
```

```python
df = data
# removing irrelevant columns like Id, Zip code
# and shifting the target variable to the last position
df = df.drop(['ID','ZIP Code','Personal Loan'],axis=1)
df.head()
```

```python
# Target variable: Personal Loan
df['Perosnal Loan'] = data['Personal Loan']
df.head()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=0.25)
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()

ann.add(Dense(units=6,activation='relu',input_dim=11))
ann.add(Dense(units=6,activation='relu'))
ann.add(Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train, epochs=30, batch_size=32)
```

```python
    y_pred = ann.predict(X_test)

    # converting to Boolean values for easier interpretation
    y_pred = (y_pred>0.5)
    y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
